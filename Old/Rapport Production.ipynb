### 1. Téléchargement Mart Submissions

```python
client = bigquery.Client(project='transformed-data')

project = client.project
query = """
    select s.*
    from mart_insurance.mart_submissions as s
    """
query_job = client.query(query)
sub = query_job.to_dataframe()
```

### 2. Traitement des submissions

```python
#Traitement des dates
sub['submission_created_at'] = pd.to_datetime(sub['submission_created_at']).dt.date
sub['bind_date'] = pd.to_datetime(sub['bind_date']).dt.date
sub['original_effective_date'] = pd.to_datetime(sub['original_effective_date']).dt.date
sub['period_start'] = pd.to_datetime(sub['period_start']).dt.date
sub['period_end'] = pd.to_datetime(sub['period_end']).dt.date
sub['policy_period_start'] = pd.to_datetime(sub['policy_period_start']).dt.date
sub['policy_period_end'] = pd.to_datetime(sub['policy_period_end']).dt.date
sub['ph_birthdate'] = pd.to_datetime(sub['ph_birthdate']).dt.date
sub['md_birthdate'] = pd.to_datetime(sub['md_birthdate']).dt.date
sub['md_license_date'] = pd.to_datetime(sub['md_license_date']).dt.date
sub['md_date_crm_50'] = pd.to_datetime(sub['md_date_crm_50']).dt.date
sub['vh_purchase_date'] = pd.to_datetime(sub['vh_purchase_date']).dt.date
sub['vh_circulation_date'] = pd.to_datetime(sub['vh_circulation_date']).dt.date
sub['vh_insurance_end_date'] = pd.to_datetime(sub['vh_insurance_end_date']).dt.date
sub.head(5)
```

```python
#Age du conducteur
sub['md_age'] = ((sub['period_start'] - sub['md_birthdate']) / np.timedelta64(1, 'Y')).apply(np.floor)
#Age du véhicule
sub['vh_age'] = ((sub['period_start'] - sub['vh_circulation_date']) / np.timedelta64(1, 'Y')).apply(np.floor)
#Ancienneté de permis
sub['md_license_duration'] = ((sub['period_start'] - sub['md_license_date']) / np.timedelta64(1, 'Y')).apply(np.floor)
```

```python
#Wakam a été appelé
sub['wakam_called'] = (sub.wakam_quote_id.notnull()).astype(int)
#A vu un tarif
sub['tariff_viewed'] = (sub.has_formulas == True).astype(int)
#A souscrit
sub['tariff_accepted'] = (sub.period_status == "Bound").astype(int)
```

### 3. Dédoublonnage et dataframes du premier & dernier appel tarifaire

```python
var_process = [
    'account_id', 
    'account_number', 
    'submission_created_at',
    'bind_date', 
    'job_id', 
    'job_number', 
    'policy_id', 
    'policy_number', 
    'tariff_viewed', 
    'wakam_called', 
    'tariff_accepted', 
    'vh_license_plate',
    'wakam_pricing_version',
    'generation_source',
    'online_offline'
    ]

var_risk = [
        'ph_email',
        'loss_history_type',
        'recommended_package',
        'package_subtype',
        'md_age',
        'md_license_duration',
        'vh_age',
        'md_prefix',
       # 'ph_gender',
        'md_crm_coefficient',
        'md_license_country',
        'vh_sra_code',
        'vh_license_plate',
        'vh_annual_km',
        'PACKAGE'
        ]        

var_premium = [
    'total_premium_rpt',
    'total_cost_rpt'
    ]
```

```python
#Liste des accounts_id pour lesquels on a plusieurs contrats souscrits
accounts_bound = sub.loc[sub['tariff_accepted'] == 1][list(var_process + var_risk + var_premium)]
accounts_bound['dup'] = accounts_bound.duplicated(subset = 'account_id', keep = False)
#Account_id en doublons
accounts_dup = accounts_bound.loc[accounts_bound['dup'] == True].sort_values(['account_id', 'submission_created_at'])
accounts_dup
#Liste des account_id dédoublonnée où il y a eu souscription
accounts_dedup = accounts_bound.drop_duplicates('account_id')
```

```python
#Ligne correspondant au premier appel tarifaire pour chaque account
subf = sub.sort_values(by = ['account_id', 'submission_created_at'], ascending = [True, True])[list(var_process + var_risk + var_premium)]
subf.drop_duplicates(['account_id'], inplace = True)
print("Il existe", subf.shape[0], "lignes lorsqu'on dédoublonne par account_id")
#Merger l'info de la contractualisation
subf = subf.merge(accounts_dedup[list(['account_id', 'tariff_accepted', 'bind_date'] + var_premium)], on = 'account_id', how = 'left', indicator = True, suffixes = ('_left', '_right'))
subf['account_bound'] = subf['tariff_accepted_right'].apply(lambda x: 1 if x == 1 else 0)
subf['bind_delay'] = (subf['bind_date_right'] - subf['submission_created_at']) / np.timedelta64(1, 'D')
subf['account_bound_0'] = ((subf.account_bound == 1).astype(int)) * ((subf.bind_delay == 0).astype(int))
subf['account_bound_3'] = ((subf.account_bound == 1).astype(int)) * ((subf.bind_delay <= 3).astype(int))
#Une variable submissions pour avoir la somme des appels tarifaires
subf['submissions'] = 1
```

```python
#Ligne correspondant au dernier appel tarifaire pour chaque account
subl = sub.sort_values(by = ['account_id', 'submission_created_at'], ascending = [True, False])[list(var_process + var_risk + var_premium)]
subl.drop_duplicates(['account_id'], inplace = True)
print("Il existe", subl.shape[0], "lignes lorsqu'on dédoublonne par account_id")
#Merger l'info de la contractualisation
subl = subl.merge(accounts_dedup[list(['account_id', 'tariff_accepted', 'bind_date'] + var_premium)], on = 'account_id', how = 'left', indicator = True, suffixes = ('_left', '_right'))
subl['account_bound'] = subl['tariff_accepted_right'].apply(lambda x: 1 if x == 1 else 0)
subl['bind_delay'] = (subl['bind_date_right'] - subl['submission_created_at']) / np.timedelta64(1, 'D')
subl['account_bound_0'] = ((subl.account_bound == 1).astype(int)) * ((subl.bind_delay == 0).astype(int))
subl['account_bound_3'] = ((subl.account_bound == 1).astype(int)) * ((subl.bind_delay <= 3).astype(int))
#Une variable submissions pour avoir la somme des appels tarifaires
subl['submissions'] = 1
```

```python
stats_subf = subf[['submissions', 'tariff_viewed', 'account_bound', 'account_bound_0', 'account_bound_3']].groupby(subf['submission_created_at']).agg(sum)
stats_subf['footprint_first'] = stats_subf['tariff_viewed'] / stats_subf['submissions']
stats_subf['conversion_first'] = stats_subf['account_bound'] / stats_subf['tariff_viewed']
stats_subf['conversion_first_0'] = stats_subf['account_bound_0'] / stats_subf['tariff_viewed']
stats_subf['conversion_first_3'] = stats_subf['account_bound_3'] / stats_subf['tariff_viewed']
stats_subf
```

```python
stats_subl = subl[['submissions', 'tariff_viewed', 'account_bound', 'account_bound_0', 'account_bound_3']].groupby(subl['submission_created_at']).agg(sum)
stats_subl['footprint_last'] = stats_subl['tariff_viewed'] / stats_subl['submissions']
stats_subl['conversion_last'] = stats_subl['account_bound'] / stats_subl['tariff_viewed']
stats_subl['conversion_first_0'] = stats_subl['account_bound_0'] / stats_subl['tariff_viewed']
stats_subl['conversion_first_3'] = stats_subl['account_bound_3'] / stats_subl['tariff_viewed']
stats_subl
```

```python
stats_conso = stats_subf.merge(
    stats_subl[['footprint_last', 'conversion_last']], how = 'left', on = 'submission_created_at')
stats_conso
```

```python
# Export final
file_path = PATH_output + 'stats_conversion2'
stats_subf.to_csv(file_path + '.csv')
```
